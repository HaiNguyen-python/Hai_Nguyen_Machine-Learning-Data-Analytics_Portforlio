```python
# Project 15: Flight Price Optimization Engine
# Goal: Analyze historical flight data to find pricing trends and predict future ticket prices using Machine Learning.
# Key Skills: Regression, Data Visualization (Seaborn), Feature Engineering.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
import warnings

# Suppress warnings for cleaner output
warnings.filterwarnings('ignore')

# STEP 1: SMART DATA GENERATION
# Simulating a dataset where prices follow specific rules:
# 1. Booking early (high 'days_until_flight') -> Lower Price.
# 2. Weekends -> Higher Price.
# 3. Evening flights -> Higher Price.

def get_flight_data():
    np.random.seed(42) # Ensure reproducible results
    n_samples = 2000
    
    airlines = np.random.choice(['AA', 'Delta', 'United'], n_samples)
    days_until = np.random.randint(1, 100, n_samples) # 1 to 100 days before flight
    dep_hours = np.random.choice([6, 9, 12, 18, 21], n_samples) # Departure times
    is_weekend = np.random.choice([0, 1], n_samples, p=[0.7, 0.3]) # 0=Weekday, 1=Weekend

    # Logic to calculate "Ground Truth" price
    prices = []
    for d, w, h in zip(days_until, is_weekend, dep_hours):
        base_price = 150
        
        # Rule 1: Dynamic Pricing (The closer to the date, the more expensive)
        # Price increases by $4 for every day closer to departure
        base_price += (100 - d) * 4 
        
        # Rule 2: Weekend Surcharge ($50 extra)
        base_price += w * 50
        
        # Rule 3: Evening Surcharge ($30 extra for flights after 6 PM)
        if h >= 18: 
            base_price += 30
            
        # Add random noise/fluctuation (+/- $20)
        base_price += np.random.randint(-20, 20)
        prices.append(base_price)

    return pd.DataFrame({
        'airline': airlines,
        'days_until_flight': days_until,
        'is_weekend': is_weekend,
        'dep_hour': dep_hours,
        'price': prices
    })

df = get_flight_data()
print("Data Generated Successfully ")
print(df.head())

```

```python
# STEP 2: DATA ANALYTICS & VISUALIZATION (EDA)
# Before Machine Learning, we analyze the data to understand business insights.

print("\n DATA ANALYTICS REPORT ")

# Insight 1: Compare Weekend vs. Weekday Prices
avg_weekend = df[df['is_weekend'] == 1]['price'].mean()
avg_weekday = df[df['is_weekend'] == 0]['price'].mean()

print(f"1. Average Weekend Price: ${avg_weekend:.2f}")
print(f"2. Average Weekday Price: ${avg_weekday:.2f}")
print(f"-> Insight: Flying on weekends is on average ${avg_weekend - avg_weekday:.2f} more expensive.")

# Insight 2: Visualize the 'Booking Window' trend
plt.figure(figsize=(10, 5))
# Scatter plot with a regression line to show the trend
sns.regplot(data=df, x='days_until_flight', y='price', 
            scatter_kws={'alpha':0.3, 'color':'blue'}, line_kws={'color':'red'})

plt.title("Price Trend: Booking Early vs. Last Minute")
plt.xlabel("Days Until Flight (Higher = Earlier Booking)")
plt.ylabel("Ticket Price ($)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

```

**+ Weekend Premium:** 

The data confirms that weekend flights are significantly more expensive (Avg: 420.11 dollars) compared to weekdays (Avg: 365.02 dollars), resulting in a ~$55 difference.

**+ Booking Trend:** 

The scatter plot reveals a strong negative correlation. The downward red trendline clearly indicates that booking earlier (higher 'Days Until Flight') drastically lowers the price, while last-minute bookings (close to 0 days) are the most expensive.

```python
# STEP 3: FEATURE ENGINEERING & PREPROCESSING
# Preparing data for the Machine Learning model.

# 1. Binning 'dep_hour' into categorical time slots
def get_time_of_day(hour):
    if hour < 12: return 'Morning'
    elif hour < 18: return 'Afternoon'
    else: return 'Evening'

df['time_of_day'] = df['dep_hour'].apply(get_time_of_day)

# 2. One-Hot Encoding
# Converting text categories (Airline, Time) into numbers (0 and 1)
df_ml = pd.get_dummies(df, columns=['airline', 'time_of_day'], drop_first=False)

# Define Features (X) and Target (y)
X = df_ml.drop(['price', 'dep_hour'], axis=1)
y = df_ml['price']

```

```python
# STEP 4: TRAIN MODEL (Random Forest)
print("\n TRAINING MODEL ")
# n_estimators=100: The model builds 100 separate decision trees and averages their predictions for better accuracy.
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X, y)
print("Model trained successfully using RandomForestRegressor.")

```

```python
# STEP 5: PREDICT NEW FLIGHT PRICES (SCENARIO TESTING)
# Simulating user queries to see how the model sets prices.

def predict_price(airline, days_until, is_weekend, time_of_day):
    # Initialize a blank dictionary matching the model's feature structure
    input_data = {col: 0 for col in X.columns}
    
    # Fill numerical features
    input_data['days_until_flight'] = days_until
    input_data['is_weekend'] = is_weekend
    
    # Fill categorical features (One-Hot Logic)
    if f'airline_{airline}' in input_data:
        input_data[f'airline_{airline}'] = 1
    if f'time_of_day_{time_of_day}' in input_data:
        input_data[f'time_of_day_{time_of_day}'] = 1
        
    # Convert to DataFrame and predict
    input_df = pd.DataFrame([input_data])
    return model.predict(input_df)[0]

print("\n PRICE PREDICTION SCENARIOS ")

# Scenario A: The Smart Planner (3 months early)
price_A = predict_price(airline='Delta', days_until=90, is_weekend=0, time_of_day='Morning')

# Scenario B: The Procrastinator (1 day before)
price_B = predict_price(airline='Delta', days_until=1, is_weekend=0, time_of_day='Morning')

# Scenario C: The 'Worst Case' (Last minute + Weekend + Evening)
price_C = predict_price(airline='Delta', days_until=1, is_weekend=1, time_of_day='Evening')

print(f"Scenario A (90 days out): ${price_A:.2f}")
print(f"Scenario B (1 day out):   ${price_B:.2f}")
print(f"Scenario C (Peak Time):   ${price_C:.2f}")

print(f"\n-> Conclusion: Booking last minute costs an extra ${price_B - price_A:.2f}.")
```
