```python
# Project 5: Spam Detection Pipeline using NLP

#Description: An end-to-end NLP pipeline that automates SMS spam detection 
#using a Naive Bayes classifier [P(Y/X) = (P(X/Y)*P(Y))/P(X)] 
#to distinguish between legitimate (Ham) and unwanted (Spam) messages.
# Libraries: requests (API), pandas (Data), sklearn (ML), seaborn (Vis)

import pandas as pd
import numpy as np
import requests
import io
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
from sklearn.metrics import confusion_matrix, classification_report

# PHASE 1: DATA ENGINEERING (Extract from API)

def fetch_spam_data():
    # URL acting as an API Endpoint (UCI Machine Learning Repository via GitHub)
    url = "https://raw.githubusercontent.com/justmarkham/pycon-2016-tutorial/master/data/sms.tsv"
    
    print("Fetching dataset from API...")
    response = requests.get(url)
    
    if response.status_code == 200:
        # Read the tab-separated values (TSV) directly from the response content
        return pd.read_csv(io.StringIO(response.content.decode('utf-8')), sep='\t', header=None, names=['label', 'message'])
    else:
        raise Exception("Failed to fetch data")

# Load data
df = fetch_spam_data()
print(f"Data Loaded Successfully. Total messages: {len(df)}")
print(df.head())

```

```python
# PHASE 2: DATA TRANSFORMATION (Preprocessing)
# 1. Convert Labels to Binary as Machine only understands numbers
# ham = 0, spam = 1
df['label_num'] = df['label'].map({'ham': 0, 'spam': 1})

# 2. Check for Class Imbalance (Important for Analytics)
print("\nClass Distribution:")
print(df['label'].value_counts())

# 3. Split Data
X = df['message']
y = df['label_num']

# Split 75% training, 25% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

```

```python
# PHASE 3: MACHINE LEARNING (NLP Pipeline)
# We build a pipeline that does two things:
# 1. CountVectorizer: Converts text into a matrix of token counts (Bag of Words)
# 2. MultinomialNB: Naive Bayes classifier (Standard for text classification)

print("\nTraining Model...")
model = make_pipeline(CountVectorizer(), MultinomialNB())
model.fit(X_train, y_train)

# Predict on Test Data
predictions = model.predict(X_test)
```

```python
# PHASE 4: DATA ANALYTICS & VISUALIZATION
# 1. Performance Metrics
print("\n Model Evaluation")
print(classification_report(y_test, predictions, target_names=['Ham', 'Spam']))

# 2. Confusion Matrix Visualization
cm = confusion_matrix(y_test, predictions)
#Confusion matrix: True Positive - True Negative - False Positive - False Negative 

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Ham', 'Spam'], yticklabels=['Ham', 'Spam'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Spam Filter Confusion Matrix')
plt.show()

# 3. Test with Custom Real-world Examples
print("\n Live Test")
custom_emails = [
    "Hey, are we still meeting for lunch today?",      # Normal
    "URGENT! You have won a free iPhone. Claim now!", # Spam
    "Your bank account security code is 1234",        # Normal
    "Cash loans with low interest. Apply quickly."    # Spam
]
custom_preds = model.predict(custom_emails)

for text, label in zip(custom_emails, custom_preds):
    result = "SPAM" if label == 1 else "Ham"
    print(f"Input: '{text}'\nResult: [{result}]\n")
```

**Interpretation of the Confusion Matrix**

**+ True Negative:** 1203 HAM messages that machine has correctly predicted

**+ True Positive:** 174 SPAM messages that machine has correctly predicted

**+ False Positive:** 4 HAM messages that machine has wrongly predicted

**+ False Negative:** 12 SPAM messages that machine has wrongly predicted


**Interpretation of the Model Evaluation**

**1. Precision (0.98):** When the model says "Spam", it is 98% sure.

**2. Recall (0.94):** The model catches 94% of all real spam messages.

**3. Accuracy (0.99):** The model is correct 99% of the time overall.

**METRICS REVIEW**

**1) PRECISION:** Accuracy of positive predictions. "When it predicts Spam, is it right?"

**2) RECALL:**    Ability to find all positives.    "Did it catch all the Spam?"

**3) F1-SCORE:**  Harmonic mean of P & R.           "The balance between Precision and Recall."

**4) SUPPORT:**   Sample size.                      "How many actual Ham/Spam messages were tested."
